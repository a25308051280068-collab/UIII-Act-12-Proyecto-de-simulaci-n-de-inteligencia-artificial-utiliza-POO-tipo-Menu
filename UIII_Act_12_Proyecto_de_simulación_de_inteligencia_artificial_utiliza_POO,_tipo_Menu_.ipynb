```python
import os

class Recepcionista:
    def __init__(self, idiomas, sistema_reservas, extension, empresa, horario, tareas):
        # Atributos de la clase
        self.idiomas = idiomas
        self.sistema_reservas = sistema_reservas
        self.extension = extension
        self.empresa = empresa
        self.horario = horario
        self.tareas = tareas

    # Función 1: Mostrar el estado/perfil de la IA en formato de tabla
    def mostrar_perfil(self):
        print("\n" + "="*50)
        print(f"      PERFIL DEL ASISTENTE DE IA: {self.empresa.upper()}")
        print("="*50)
        # Formateo manual tipo tabla con alineación
        print(f"{'Atributo':<25} | {'Valor Actual':<25}")
        print("-" * 50)
        print(f"{'Idiomas hablados':<25} | {', '.join(self.idiomas):<25}")
        print(f"{'Sistema de reservas':<25} | {self.sistema_reservas:<25}")
        print(f"{'Extensión telefónica':<25} | {self.extension:<25}")
        print(f"{'Empresa/Sede':<25} | {self.empresa:<25}")
        print(f"{'Horario de entrada':<25} | {self.horario:<25}")
        print(f"{'Tareas administrativas':<25} | {', '.join(self.tareas):<25}")
        print("="*50 + "\n")

    # Función 2: Simular la atención telefónica (IA procesando datos)
    def atender_llamada(self, idioma_usuario, consulta):
        print("\n--- [📞 LLAMADA ENTRANTE] ---")
        if idioma_usuario.lower() in [i.lower() for i in self.idiomas]:
            print(f"🤖 IA: 'Bienvenido a {self.empresa}. Tomando datos...'")
            print(f"🧠 IA Procesando: Redirigiendo consulta '{consulta}' a la extensión {self.extension}...")
            print("✨ ¡Consulta procesada con éxito por el módulo de IA!")
        else:
            print(f"❌ Error de IA: El idioma '{idioma_usuario}' no está instalado en mi sistema.")
            print(f"🌐 Idiomas soportados actualmente: {', '.join(self.idiomas)}")

    # Función 3: Gestionar el sistema de reservas
    def gestionar_reserva(self, nombre_cliente, hora_reserva):
        print("\n--- [📅 MÓDULO DE RESERVAS] ---")
        print(f"🔄 Conectando con el sistema: {self.sistema_reservas}...")
        print(f"✅ Reserva guardada con éxito para {nombre_cliente} a las {hora_reserva}.")
        print(f"🕒 Nota: Recuerde que mi horario de atención inicia a las {self.horario}.")

    # Función 4: Actualizar las tareas de la IA
    def actualizar_tareas(self, nueva_tarea):
        self.tareas.append(nueva_tarea)
        print(f"\n⚙️ [SISTEMA OPTIMIZADO]: Se ha añadido '{nueva_tarea}' a mis tareas administrativas.")


# --- FLUJO PRINCIPAL / MENÚ INTERACTIVO ---

# Instanciamos el objeto 'ia_recepcionista' con datos iniciales
ia_recepcionista = Recepcionista(
    idiomas=["Español", "Inglés"],
    sistema_reservas="OpenTable AI v2.5",
    extension="Ext. 404",
    empresa="CyberCore Tech",
    horario="08:00 AM",
    tareas=["Filtrar correos", "Agendar juntas"]
)

# Bucle del Menú (Sin bloques Try/Except, validación directa)
ejecutando = True

while ejecutando:
    print("🤖 *** SIMULADOR DE IA: RECEPCIONISTA VIRTUAL *** 🤖")
    print("1. Ver perfil y atributos de la IA (Formato Tabla)")
    print("2. Simular atención de llamada")
    print("3. Registrar una nueva reserva")
    print("4. Agregar tarea administrativa a la IA")
    print("5. Salir")

    opcion = input("Selecciona una opción (1-5): ")

    if opcion == "1":
        ia_recepcionista.mostrar_perfil()

    elif opcion == "2":
        idioma = input("¿En qué idioma vas a hablar? (Ej: Español, Inglés, Francés): ")
        motivo = input("¿Cuál es el motivo de tu llamada?: ")
        ia_recepcionista.atender_llamada(idioma, motivo)

    elif opcion == "3":
        cliente = input("Nombre del cliente: ")
        hora = input("Hora de la reserva (Ej: 09:00 AM): ")
        ia_recepcionista.gestionar_reserva(cliente, hora)

    elif opcion == "4":
        nueva_tarea = input("Escribe la nueva tarea administrativa para la IA: ")
        ia_recepcionista.actualizar_tareas(nueva_tarea)

    elif opcion == "5":
        print("\nApagando sistema de IA... ¡Hasta luego!")
        ejecutando = False

    else:
        print("\n❌ Opción no válida. Por favor, elige un número del 1 al 5.\n")

    # Pausa estética para que los alumnos puedan leer el resultado antes de limpiar pantalla
    if ejecutando:
        input("\nPresiona Enter para volver al menú...")
        # Limpia la pantalla en Windows o Linux/Mac para mantenerlo ordenado
        os.system('cls' if os.name == 'nt' else 'clear')
```
